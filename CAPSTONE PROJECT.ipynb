```python
missing_val_features=[each for each in df.columns if df[each].isnull().sum()>1]
#printing features that have missing values
print(missing_val_features)
#for loop to interate through features and plotting a bar chart
for feature in missing_val_features:
    height=df[feature].isnull().sum()
    plt.bar(feature,height)
```

# # Handeling Outliers 


```python
Q1=dff.quantile(0.25).round(3)
Q3=dff.quantile(0.75).round(3)
IQR=Q3-Q1
print(IQR)
```

```python
dfout= dff[~((dff < (Q1 - 1.5 * IQR)) | (dff > (Q3 + 1.5 * IQR))).any(axis = 1)] # removing outliers from whole data set
dfout.shape
```

```python
from sklearn.preprocessing import StandardScaler #data normalization with sklearn
sc=StandardScaler()
x_train=sc.fit_transform(x_train) #fit scaler on training data
x_test=sc.transform(x_test) #transform testing data
```

# MACHINE LEARNING TECHNIQUES

```python
#Importing  DecisionTree Regressor from sklearn.tree library
from sklearn.tree import DecisionTreeRegressor
#calling decision tree regressor with max_depth as 3 and calling it to dt
dt=DecisionTreeRegressor(max_depth=3)
```

```python
#fitting X_train ,Y_train into dt
dt.fit(x_train,y_train)
```

```python
#score of train model
print(dt.score(x_train,y_train))
#score of test model
print(dt.score(x_test,y_test))
```

```python
# by using dt.predict function on x_test and calling it to y_predict to predict a value
y_pred=dt.predict(x_test)
#the code here provides is making use of predict function to generate predictions based on the input data x_test
```

```python
print(y_pred)# checking the predict values
```

```python
# importing metrices from sklearn library
from sklearn import metrics
#metrix gives r2 scores to cross verify accuracy of model
```

```python
# checking the r2 scores to y_test,y_predict to predict the accuracy.
r_square=metrics.r2_score(y_test,y_pred)
print("R square error for decision tree regressor is:",round(100*(r_square),3))
```

```python
#by using dt.predict function on x_test and calling it to y_predict to predict a value
y_predict=dt.predict(x_test)
#importing metrics from sklearn library
from sklearn import metrics
#metrics gives r2 score to cross verify accuracy of model
r_square=metrics.r2_score(y_test,y_predict)
print("R-Square eror for decision tree regressor is :",round(100*(r_square),2))
```

```python
# importing random forest regressor from sklearn.ensemble and it rfr with random_state 5 and fitting it to x_train,y_train
from sklearn.ensemble import RandomForestRegressor
#importing metrices from sklearn library
from sklearn import metrics
rfr=RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
```

```python
# by using (rfr.predict function x)test and calling it to y_predict to predict a value
y1_pred=rfr.predict(x_test)
```

```python
#calculating mean absolute error,mean squared error,root mean squared error,mean absolute percentage error and accuracy on y_test
print("Mean Absolute Error (MAE):",metrics.mean_absolute_error(y_test,y1_pred))
print("Mean Squared Error (MSE):",metrics.mean_squared_error(y_test,y1_pred))
# while both MAE/MSE are performance measures

print("Root Mean Squared Error (RMSE:)"),
np.sqrt(metrics.mean_squared_error(y_test,y1_pred))
# root mean squared error is a measure of how well the machine learning model can perform.

mape=np.mean(np.abs((y_test-y1_pred)/np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(100*(1-mape),2)

print('Mean Absolute Percentage Error (MAPE):',round(mape*100,2))
print("Accuracy:",round(100*(1-mape),2))
```
